```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Đọc tệp đã sửa từ Google Drive
file_path_fixed = '/content/drive/My Drive/dataset_cho_python1.csv'
data = pd.read_csv(file_path_fixed)

# Chuyển đổi dữ liệu nếu cần thiết
data['Rating Average'] = pd.to_numeric(data['Rating Average'].str.replace(',', '.'), errors='coerce')
data['Complexity Average'] = pd.to_numeric(data['Complexity Average'].str.replace(',', '.'), errors='coerce')

# Hàm tạo biểu đồ

def generate_column_charts(data):
    for column in data.columns:
        plt.figure(figsize=(10, 6))
        if data[column].dtype in ['float64', 'int64']:
            sns.histplot(data[column].dropna(), kde=True, bins=30, color='blue')
            plt.title(f"Distribution of {column}")
            plt.xlabel(column)
            plt.ylabel("Frequency")
        elif data[column].dtype == 'object':
            value_counts = data[column].value_counts().head(10)  # Hiển thị top 10 giá trị phổ biến
            value_counts.plot(kind='bar', color='orange')
            plt.title(f"Top 10 Categories in {column}")
            plt.xlabel(column)
            plt.ylabel("Count")
        plt.tight_layout()
        plt.show()

# Gọi hàm tạo biểu đồ
generate_column_charts(data)

print("Biểu đồ đã được hiển thị trực tiếp.")
```
